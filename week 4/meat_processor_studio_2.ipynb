## 3.2 From your observation where are the most false positives coming from? How you can fix it?

Most false positives coming from the Class 0 being predicted as Class 2 and vice versa. In order to fix the problem, we need to improve the accuracy by:
* Oversampling/Undersampling the majority/minority class 
* model-tunning hyperparameters to reduce overfitting or underfitting
* use multiple models to reduce the weaknesses of individual models, 
* use SMOTE (Synthetic Minority Over-sampling Technique)
