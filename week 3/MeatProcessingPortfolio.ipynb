# Step 2: Create Composite columns

Create some composite data points for your data as follows
1) Root mean square value of x and y
2) Root mean square value of y and z
3) Root mean square value of z and x
4) Root mean square value of x, y and z
5) Roll `(180 * atan2(accelY, sqrt(accelX * accelX + accelZ * accelZ))/PI)` 
6) Pitch `(180 * atan2(accelX, sqrt(accelY * accelY + accelZ * accelZ))/PI)`

So, you will get 12 more columns for 2 settings from the above. Merge this with your data in Step 1. Now you will have data file with 20 columns.

## Column 8-13: 6 computed columns in step 2 for column Set 1

```python
# 1 Root mean square value of x and y
df_combined_data['rms_xy_set_1'] = np.sqrt(np.mean(df_combined_data['Right Upper Arm x']**2)) + np.sqrt(np.mean(df_combined_data['Right Upper Arm y']**2))
# 2 Root mean square value of y and z
df_combined_data['rms_yz_set_1'] = np.sqrt(np.mean(df_combined_data['Right Upper Arm z']**2)) + np.sqrt(np.mean(df_combined_data['Right Upper Arm y']**2))
# 3 Root mean square value of x and z
df_combined_data['rms_zx_set_1'] = np.sqrt(np.mean(df_combined_data['Right Upper Arm z']**2)) + np.sqrt(np.mean(df_combined_data['Right Upper Arm x']**2))
# Root mean square value of x, y and z
df_combined_data['rms_xyz_set_1'] = np.sqrt(np.mean(df_combined_data['Right Upper Arm x']**2)) + np.sqrt(np.mean(df_combined_data['Right Upper Arm y']**2)) + np.sqrt(np.mean(df_combined_data['Right Upper Arm z']**2))

# Roll
df_combined_data['rms_roll_set_1'] = 180 * np.arctan2(df_combined_data['Right Upper Arm y'], np.sqrt(df_combined_data['Right Upper Arm x']**2 + df_combined_data['Right Upper Arm z']**2)) / np.pi
# Pitch
df_combined_data['rms_pitch_set_1'] = 180 * np.arctan2(df_combined_data['Right Upper Arm x'], np.sqrt(df_combined_data['Right Upper Arm y']**2 + df_combined_data['Right Upper Arm z']**2)) / np.pi
```

# Step 3: Data pre-processing and Feature computation

Now you will need to write code to create statistical features per minute (1 minute = consecutive 60 frames) from the 18 columns. Like we use in Studio 3 compute the following features for each column 2-19

```python
from scipy.signal import find_peaks
from scipy.integrate import simpson as simps
```

```python
# Read the updated data
df_updated_data = pd.read_csv('updated_combined_data.csv')
```

```python
# number of frame per second
frame_per_mins = 60

# List all columns, except first and last column
columns_without_first_and_last = updated_data.columns[1:-1]

feature_rows = []

# Process each column in frame
for start in range(0, len(df_updated_data), frame_per_mins):
    if start + frame_per_mins > len(df_updated_data):
        break

    segment = df_updated_data.iloc[start:start + frame_per_mins]

    features = {}

    for column in columns_without_first_and_last:
        col_data = segment[column]

        features[column + '_mean'] = col_data.mean()
        features[column + '_std'] = col_data.std()
        features[column + '_min'] = col_data.min()
        features[column + '_max'] = col_data.max()

        # Calculate AUC
        features[column + '_auc'] = simps(col_data) # use last value

        # Calculate Peaks
        features[column + '_peaks'] = len(find_peaks(col_data)[0])

        # Append the class label
        features['class'] = segment['class'].values[0]
        
        # Append the features into the list
        feature_rows.append(features)

feature_df = pd.DataFrame(feature_rows)


# Save into new csv file
feature_df.to_csv('feature_computation_data.csv', index=False)
        
```
